```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd 
df=pd.read_csv("recent_100_vulnerabilities.csv")
```

```python
top_vendors = df['vendorProject'].value_counts().head(10)

plt.figure(figsize=(10, 6))
sns.barplot(y=top_vendors.index, x=top_vendors.values, palette='crest')
plt.title("Top 10 Vendors with Most Vulnerabilities")
plt.xlabel("Vulnerability Count")
plt.ylabel("Vendor")
plt.tight_layout()
plt.show()
```
