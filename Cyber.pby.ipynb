```python
#Known Exploited Vulnerabilities Catalog
```

```python
import pandas as pd
import numpy as np 
```

```python
df=pd.read_csv(f"known_exploited_vulnerabilities.csv")
```

```python
df.info()
```

```python
df.head(5)
```

```python
#Hindale Missing Value 
```

```python
print(df.isnull().sum())
```

```python
df = df.dropna(subset=['cveID', 'dateAdded'])
```

```python
df.head(5)
```

```python
df['notes'] = df['notes'].fillna("No notes provided")
```

```python
# Clean Date Columns
```

```python
date_cols = ['dateAdded', 'dueDate']
for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')
```

```python
df.head(5)
```

```python
# Normalize Text Columns
```

```python
df['vendorProject'] = df['vendorProject'].str.strip().str.lower()
df['vulnerabilityName'] = df['vulnerabilityName'].str.strip().str.title()
```

```python
df['vulnerabilityName'] = df['vulnerabilityName'].replace({
    "Windows Rpc Vulnerability": "Windows RPC Vulnerability",
   
})
```

```python
df.head()
```

```python
#Remove Duplicates
```

```python
df = df.drop_duplicates()
```

```python
df = df.drop_duplicates(subset=['cveID'])
```

```python
df.head()
```

```python
#Feature Engineering
```

```python
df['year_added'] = df['dateAdded'].dt.year
df['month_added'] = df['dateAdded'].dt.month
```

```python
df['days_since_added'] = (pd.Timestamp.today() - df['dateAdded']).dt.days
```

```python
df.head()
```

```python
#Filter Relevant Records
```

```python
df = df[df['requiredAction'].notnull()]
```

```python
df.info()
```

```python
#Final Cleaned DataFrame
```

```python
df.to_csv("cleaned_cyber_vulnerabilities.csv", index=False)
```
